```python
df['Month'] = df.index.month  
monthly_counts = df.groupby(['Month', 'Casualty_severity']).size().unstack()

plt.figure(figsize=(12, 6))
monthly_counts.plot(kind='bar', stacked=True)
plt.title('Casualty Severity Over Time')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(title='Casualty Severity')
plt.show()
```

```python
monthly_mean = df.resample('M').size().mean(level=0)
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_mean.index.month_name(), y=monthly_mean)
plt.title('Average Accidents per Month')
plt.xlabel('Month')
plt.ylabel('Average Number of Accidents')
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(resampled_df['Day_of_week'], order=(1, 1, 1))
fit_model = model.fit()
forecast = fit_model.get_forecast(steps=30)  
forecast_ci = forecast.conf_int()

plt.figure(figsize=(12, 6))
plt.plot(resampled_df['Day_of_week'], label='Actual')
plt.plot(forecast.predicted_mean, color='red', label='Forecast')
plt.fill_between(forecast_ci.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='pink', alpha=0.2)
plt.title('ARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Number of Accidents')
plt.legend()
plt.show()
```
